```python
import sys
import os
if "/opt/condaenv/anaconda2/envs/sentimentInduction" not in sys.path:
    sys.path.append("/opt/condaenv/anaconda2/envs/sentimentInduction")
if "/opt/condaenv/anaconda2/envs/sentimentInduction/lib/python2.7/site-packages" not in sys.path:
    sys.path.append("/opt/condaenv/anaconda2/envs/sentimentInduction/lib/python2.7/site-packages")
import pandas as pd
from socialsentRun.socialconfig import config
COLLECT_ACCURACIS_DIR = config.get("COLLECT_ACCURACIS_DIR")
import ujson
open_file = os.path.join(COLLECT_ACCURACIS_DIR,"food_evaluations_by_params.txt")
dfv = pd.read_json(open_file,orient='records',lines=True,encoding="utf-8")
dfv.head()
```

```python
dfv = dfv.set_index(['emb_params','method'])
dfv.head()
```

```python
dfmeanallvocabs = dfv.copy()
```

```python
dfmeanallvocabs["mean_accuracy_diff_vocab_sizes"] = dfmeanallvocabs.groupby(['emb_params','method'])['accuracy'].mean()
dfmeanallvocabs['var_diff_vocab_sizes'] = dfmeanallvocabs.groupby(['emb_params','method'])['accuracy'].var()
dfmeanallvocabs.drop(['accuracy'],inplace=True,axis=1)
dfmeanallvocabs.head()
```

```python
# dfmeanallvocabs_stacked.unstack(0)
```

```python
dfmeanallvocabs = dfmeanallvocabs.reset_index()
dfmeanallvocabs = dfmeanallvocabs.drop_duplicates()
len(dfmeanallvocabs)
```

```python
dfmeanallvocabs.head()
```

```python
dfmeanallvocabs = dfmeanallvocabs.set_index(['emb_params','method'])
dfmeanallvocabs.drop('var_diff_vocab_sizes',axis=1,inplace=True)
dfmeanallvocabs.head()
```

```python
dfmeanallvocabs.unstack('method') # Note for paper
```

```python
dfmeanallvocabs.to_html("mean_accuracies_for_all_methods_types_and_training_param_type.html")
```

```python
dfmeanallmethodsperparam = dfmeanallvocabs.copy()
dfmeanallmethodsperparam = dfmeanallmethodsperparam.set_index(['emb_params'])
dfmeanallmethodsperparam.head()
```

```python
dfmeanallmethodsperparam["mean_accuracy_all_ind_methods"] = dfmeanallmethodsperparam.groupby(['emb_params'])['mean_accuracy_diff_vocab_sizes'].mean()
dfmeanallmethodsperparam['var_all_ind_methods'] = dfmeanallmethodsperparam.groupby(['emb_params'])['mean_accuracy_diff_vocab_sizes'].var()
dfmeanallmethodsperparam.drop(['mean_accuracy_diff_vocab_sizes','var_diff_vocab_sizes'],axis=1,inplace=True)
dfmeanallmethodsperparam.head()
```

```python
dfmeanallmethodsperparam = dfmeanallmethodsperparam.reset_index()
dfmeanallmethodsperparam.head()
```

```python
dfmeanallmethodsperparam = dfmeanallmethodsperparam.drop_duplicates()
dfmeanallmethodsperparam # Note for paper
```

```python
dfmeanallmethodsperparam['mean_of_mean'] = dfmeanallmethodsperparam['mean_accuracy_all_ind_methods'].mean()
```

```python
dfmeanallmethodsperparam['var_of_mean'] = dfmeanallmethodsperparam['mean_accuracy_all_ind_methods'].var()
dfmeanallmethodsperparam
```
